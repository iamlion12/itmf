```python
import numpy as np
import imageio
import os
from ast import literal_eval
import matplotlib.pyplot as plt
import pickle

%matplotlib inline
```

```python
with open("get_labels_info", 'r') as File:
    labels = literal_eval(File.read())
```

```python
images = [[imageio.imread('./resized/'+img) for img in labels[key] if os.path.isfile('./resized/'+img)] for key in labels.keys()]
```

```python
y = np.vstack((np.full((len(images[0]), 1), 0),
               np.full((len(images[1]), 1), 1),
               np.full((len(images[2]), 1), 1)))
```

```python
x = np.vstack((np.array(images[0]), np.array(images[1]), np.array(images[2])))
```

```python
index = 3501

plt.imshow(x[index])
plt.legend()
plt.show()
```

```python
np.save('x.npy', x)
np.save('y.npy', y)
```
